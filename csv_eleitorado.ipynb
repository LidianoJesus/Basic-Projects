# Analise do eleitorado brasileiro

```python
# Carregando as bibliotecas
import pandas as pd
%matplotlib inline
```

```python
# Carregando o arquivo csv
df = pd.read_csv('eleitorado_municipio_2020.csv', sep=';', encoding = 'latin1')

# Verificando as cinco primeiras linhas do DataFrame - orientação por linhas
df.head()
```

```python
# Verificando as cinco primeiras linhas do DataFrame - orientação por colunas
df.head().T
```

```python
# Verificando o tipo e a quantidade de dados
df.info()
```

```python
# Qual cidade tem o maior numero de PCD´s 
df.nlargest(5, 'QTD_ELEITORES_DEFICIENTE')
```

```python
# Sumariação de eleitores
print('Eleitoras:', df['QTD_ELEITORES_FEMININO'].sum())
print('Eleitores:', df['QTD_ELEITORES_MASCULINO'].sum())
print('Não informado:', df['QTD_ELEITORES_NAOINFORMADO'].sum())
```

```python
# Criação de variaveis para calcular os percentuais por grupo
tot_eleitores = df['QTD_ELEITORES'].sum()
tot_fem = df['QTD_ELEITORES_FEMININO'].sum()
tot_masc = df['QTD_ELEITORES_MASCULINO'].sum()
tot_nao = df['QTD_ELEITORES_NAOINFORMADO'].sum()
```

```python
# Mostrar os valores percentualmente1
print('Eleitoras:', tot_fem / tot_eleitores)
print('Eleitores:', tot_masc / tot_eleitores)
print('Não informado:', tot_nao / tot_eleitores)
```

```python
# Mostrar os valores percentualmente2
print('Eleitoras:', ((tot_fem / tot_eleitores) * 100).round(2))
print('Eleitores:', ((tot_masc / tot_eleitores) * 100).round(2))
print('Não informado:', ((tot_nao / tot_eleitores) * 100).round(2))
```

```python
# Quantos municipios tem mais homens do que mukheres?
df[df['QTD_ELEITORES_MASCULINO'] > df['QTD_ELEITORES_FEMININO']].count()
```

```python
# Criando uma coluna para indicar uma relação entre os generos feminimo e marculino
df['RELACAO_FM'] = df['QTD_ELEITORES_FEMININO'] / df['QTD_ELEITORES_MASCULINO']
```

```python
# Quais são os municipios com mais relação entre os generos feminino e masculino
df.nlargest(5, 'RELACAO_FM').T
```

```python
# Quais são os municipios com menos relação entre os generos feminino e masculino
df.nsmallest(5, 'RELACAO_FM').T
```

```python
# Criação de DataFrame somente com municipios brasileiros
# Isso foi necessário porque se percebeu que na listagem original existiam municipios com eleitores no exterior 
# o copy significa que fizemos um mecanismo de proteção ao DataFrame original, copiando e fazendo uma cópia
df_brasil = df[df['CD_PAIS'] == 1].copy()

df_brasil.head(3).T
```

```python
# Quais são os municipios com mais relação entre os generos feminino e masculino
df_brasil.nlargest(5, 'RELACAO_FM').T
```

```python
# Quais são os municipios com menos relação entre os generos feminino e masculino
df_brasil.nsmallest(5, 'RELACAO_FM').T
```

```python
# Plotando um grafico de distrfibuição da relação f/m
df_brasil['RELACAO_FM'].plot.hist()
```

```python
# Plotando um grafico de distrfibuição da relação f/m
# Aumentando a visibilidade de fatores
df_brasil['RELACAO_FM'].plot.hist(bins = 100)
```

```python
# Importando uma outra biblioteca de graficos
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plotando a distribuição da relação entre os generos feminimo e masculino
# Este grafico apresenta também uma curva de aproximação
sns.distplot(df_brasil['RELACAO_FM'], bins = 100, color = 'red')
```

```python
# XXX
sns.distplot(df_brasil['RELACAO_FM'], bins = 100, color = 'red')

plt.title('Relação Eleitoras/Eleitores', fontsize = 18)
plt.xlabel('Eleitoras/Eleitores', fontsize = 14)
plt.ylabel('Frequência', fontsize = 14)
plt.axvline(1.0, color = 'black', linestyle = '--')
```

```python
# Verificando a distribuição entre os generos, porém mostrando os municipios
# Cada ponto repesentado é um municipio
sns.swarmplot(data = df_brasil, x = 'NM_REGIAO', y = 'RELACAO_FM')
```

```python
# Verificando a distribuição entre os generos, porém mostrando os municipios
# 
sns.swarmplot(data = df_brasil, x = 'NM_REGIAO', y = 'RELACAO_FM')

plt.axhline(1.0, color = 'black', linestyle = '--')
```
