```python
# Request é uma biblioteca para requisição https
import requests

# BSoup é uma lib para scrapping
from bs4 import BeautifulSoup

# Lib para dados
import pandas as pd

# Libs para Datas e Meses
import datetime
from dateutil.relativedelta import relativedelta
```

```python
# Carregando a pagina

# Site

Site = 'http://www.ssp.sp.gov.br/Estatistica/ViolenciaMulher.aspx'

# Carregar a pagina pelo request
Pagina = requests.get(Site)
```

```python
# Coletar as informações da pagina

Coleta = BeautifulSoup(Pagina.text, 'html.parser')
```

```python
# Verificar o titulo
print(Coleta.title)
```

```python
# Coletando as tabelas
Tabelas = Coleta.find_all('table', attrs={ 'class' :  'table table-striped table-hover table-condensed' } )

# Verificar as tabelas
print('Localizando: ', len(Tabelas), 'tabelas')
```

```python
# Listas para informação
Rotulos = []
Dados = []
Periodos = []

# Primeiro registro
Registro_Inicial = datetime.datetime(2021, 9, 1)

# Variavel de apoio para o Loop´

Loop = 0

# Loop para minerar os dados
for Consulta in Tabelas:

  # Filtrar a linha da tela
  Linha = Consulta.find_all('tr')[2]

  # Filtrar o rotulo
  Rotulo = Linha.find_all('td')[0].text

  # Informação
  Informação = Linha.find_all('span')[0].text

  # Calcular a data
  Data = Registro_Inicial - relativedelta(months=Loop)

  # Caso seja diferente de feminicidio

  if Rotulo == 'FEMINICÍDIO':
    Rotulos.append(Rotulo)
    Dados.append(Informação)
    Periodos.append(Data)

  else:
      pass

 # print(Informação)

  Loop += 1
```

```python
# Organizar os dados

Dicionario = {
    'Indicador' : Rotulos,
    'Quantidade' : Dados,
    'Periodo' : Periodos
}

# Criar uma base de dados

Base_Dados = pd.DataFrame( Dicionario )
```
