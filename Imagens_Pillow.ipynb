```python
# Biblioteca para trabalhar com matrizes
import numpy as np

# Biblioteca para trabalhar com imagens
from PIL import Image
```

```python
# Carregando uma imagem

# Definindo o nome da imagem
Foto = 'Logo.jpg'

# Carregando a imagem
Imagem = Image.open( Foto )
Imagem
```

```python
# Verificar o Tamanho
Imagem.size
```

```python
# Veriricar a largura
Imagem.width
```

```python
# Verificar a altura
Imagem.height
```

```python
# Retornando a imagem coomo objeto em bytes
Imagem.tobytes()
```

```python
# Extrair os dados associado a imagem
# Esse dicionrio  usado por manipuladores de arquivos
Imagem.info
```

```python
# Converter a imagem para um formato de Matriz 

# Bibliotecas para mostrar as imagens
from matplotlib import image
from matplotlib import pyplot

# Cada linha ir representar RGB
# R = Red ( Vermelhro )
# G = Green ( Verde )
# B = Blue ( Azul )

# Lerdo a imagem como Array
Imagem_Formato_Array = image.imread('Logo.jpg')
Imagem_Formato_Array
```

```python
# Verificando a dimenso do Array Imagem
Imagem_Formato_Array.shape
```

```python
# Plotar a Imagem dentro de um Grfico
# Esse tipo de situao  muito usado para ML com Imagens

# Mostrando a imagem em um grfico
pyplot.imshow(Imagem_Formato_Array)
```

```python
# Ajustando o Tamanho da Imagme

# Definindo a dimenso
Redimenso = ( 250, 250 )

# Redimensionando a imagem
Imagem = Imagem.resize( size=Redimenso )
Imagem
```

```python
# Transformando em Miniatura

# Definindo tamanho
Tamanho = (50,50)

# Aplicando a funo para miniatura
Imagem.thumbnail(size=Tamanho)
```

```python
Imagem
```

```python
# Rotao funcina no sentindo horrio

# 45 Graus
Graus_01 = 45

# 90 Graus
Graus_02 = 90

# 180 Graus
Graus_03 = 180

# Rotacionando 45 Graus 
Imagem.rotate( angle=Graus_01 )
```

```python
# Rotacionando 90 Graus 
Imagem.rotate( angle=Graus_02 )
```

```python
# Rotacionando 180 Graus 
Imagem.rotate( angle=Graus_03 )
```

```python
# Podemos transformar a imagem com o metodo 'Transpose'

# Opes do transpose
# PIL.Image.FLIP_LEFT_RIGHT
# PIL.Image.FLIP_TOP_BOTTOM
# PIL.Image.ROTATE_90
# PIL.Image.ROTATE_180
# PIL.Image.ROTATE_270
# PIL.Image.TRANSPOSE
# PIL.Image.TRANSVERSE

# Rotacionando
Imagem.transpose( method=Image.ROTATE_90 )
```

```python
# Rotacionando usando o metodo transpose
Imagem.transpose( method=Image.TRANSVERSE )
```

```python
# Divir uma imagem em tres novas imagens
# cada uma contendo uma cpia de uma das faixas originais 
# (vermelho, verde, azul)

# Aplicando a Diviso das Imagens
Imagens = Image.Image.split(Imagem )
```

```python
# Faixa de Vermelho
Imagens[0]
```

```python
# Faixa de Verde
Imagens[1]
```

```python
# Faixa de Azul
Imagens[2]
```

```python
# Diminuindo o numero de cores de uma imagem

# Definindo que maximo de 50 cores
Maximo_Cores = 50

# Aplicando o metodo
Imagem.quantize( colors=Maximo_Cores )
```

```python
# Metodo Reduce

# Definindo a posio para ser reduzida
Posicoes = ( 10, 40, 150, 150 )

# Aplicando a funo
Imagem.reduce( factor=1, box=Posicoes  )
```

```python
# Converter as imagens em escalas

# L = Cinza
# RGB = 'Red, Green, Blue'
# CMYK = 'Cyan', 'Magenta', 'Yellow', 'Black'

# Convertendo a imagem para escalas de cinza
Imagem.convert(mode='L')
```

```python
# Salvar a imagem que fez o ajuste
Imagem.save('Minerando_Imagens.jpg')
```
