```python
# Biblioteca para minera audios
import librosa
import librosa.display

# Biblioteca para analises gráficas
import matplotlib.pyplot as plt
```

```python
# Vamos utilizar um audio da propria Lib
# Tu pode usar qualquer auido

# Usando audio exemplo da Librosa
Audio = librosa.example('nutcracker')

# Carregnado o audio
Matriz, Som = librosa.load( Audio )
```

```python
# Conversão do audio para o formato de matriz
# Cada valor representa uma onda
Matriz
```

```python
# Verificando a dimensão dos dados de ondas
Matriz.shape
```

```python
# taxa de amostragem do som
Som
```

```python
# Podemos mudar o comportamento do audio para 44,1 kHz.
# Podemos passar parametros do KHz
# Será retornado uma matriz com os valores 

librosa.load( Audio, sr=44100)
```

```python
# Plotar a matriz e entender a escala do som
# Definindo o tamanho do gráfico
plt.figure(figsize=(14, 5))
# Passando a matriz como parametro para o plot do Audio
librosa.display.waveplot( Matriz, sr=Som );
```

```python
# Vamos alterar a frequência de amostragem
# Veja que o tempo do audio mudou

# Definindo o tamanho do gráfico
plt.figure(figsize=(14, 5))
# Passando a matriz como parametro para o plot do Audio
librosa.display.waveplot( Matriz, sr=( Som * 3 ) );
```

```python
# Plotando um Spetograma

# O STFT representa um sinal no domínio da frequência do tempo, 
# calculando as transformadas discretas de Fourier (DFT) em 
# janelas de sobreposição curtas.
# em outras palavras é uma derivada
Sinal = librosa.stft( Matriz )

# Converter um espectrograma de amplitude em espectrograma em escalas de decibel
Decibeis = librosa.amplitude_to_db( abs(Sinal) )

# Definindo o tamanho do gráfico
plt.figure(figsize=(14, 5))

# Plotando o Espectrograma
librosa.display.specshow( 
    # Nossa matriz com os decibeis
    Decibeis, 
    # Taxa de amostragem
    sr=Som, 
    # Defiindo o eixo x ( Tempo do audio )
    x_axis='time', 
    # Defiindo o eixo y ( Hertz )
    y_axis='hz')

# barra de cores para o gráfico
plt.colorbar();
```

```python
# O retorno da frequência do tempo
Sinal[0:2]
```

```python
# O retorno dessa função é uma matriz com diversos valores
Decibeis[0:2]
```

```python
# Tipos de Frequência

# 'linear', 'fft', 'hz': a faixa de frequência é determinada pela janela FFT e a taxa de amostragem.
# 'log': o espectro é exibido em uma escala logarítmica.
# 'fft_note': o espectro é exibido em uma escala logarítmica com as alturas marcadas.
# 'fft_svara': o espectro é exibido em uma escala logarítmica com svara marcado.
# 'mel': as frequências são determinadas pela escala de mel.
# 'cqt_hz': as frequências são determinadas pela escala CQT.
# 'cqt_note': os tons são determinados pela escala CQT.
# 'cqt_svara': como cqt_note, mas usando Hindustani ou Carnatic svara

# Podemos mudar a frequência de Hertz para outras 

# Definindo o tamanho do gráfico
plt.figure(figsize=(14, 5))

# Plotando o Espectrograma
librosa.display.specshow( 
    # Nossa matriz com os decibeis
    Decibeis, 
    # Amostra de frequencia
    sr=Som, 
    # Defiindo o eixo x ( Tempo do audio )
    x_axis='time', 
    # Defiindo o eixo y ( Frequencia em escala logarítmica )
    y_axis='log')

# barra de cores para o gráfico
plt.colorbar();
```

```python

# Apare o silêncio inicial e final 
yt ,  index  =  librosa.effects.trim ( Matriz ) 
# Imprime as durações 
print ( librosa . get_duration ( Matriz ),  librosa . get_duration ( yt )) 
```
